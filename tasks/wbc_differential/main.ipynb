About
=====

This task is for performing a White Blood Cell differential.
It works in two stages. In the first stage, a CNN is executed
across the input image, in tiles, and predicts the centroid
heatmap for each whiteblood cell in the frame. After that, the
output frame is converted to a series of local peaks, where each
peak represents the centroid of a white blood cell. Finally, for
each peak, a tile cropped around it and passed to a classification
network that determines what the class of white blood cell the peak
belongs to. There is also a null class, for false positives from the
first network.
